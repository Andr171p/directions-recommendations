```python
import pandas as pd
```

```python
file_path = r"C:\Users\andre\IdeaProjects\TyuiuDirectionsRecSys\datasets\tyuiu_students_with_ids_2019-2024.csv"

df = pd.read_csv(file_path)

df.head(3)
```

```python
unique_df = df[["Направление подготовки", "ID"]].drop_duplicates()

unique_df
```

```python
unique_df.to_csv("tyuiu_unique_directions.csv")
```

```python
import json

with open("Структурированные_направления_подготовки.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
def get_id_by_direction_name(direction_name):
    filtered_rows = unique_df[unique_df['Направление подготовки'] == direction_name]
    if len(filtered_rows) > 0:
        return filtered_rows['ID'].values[0]
    else:
        return None 
```

```python
for direction in data:
    direction_id = get_id_by_direction_name(direction["Шифр и наименование направления"])
    direction["id"] = int(direction_id) if direction_id else -1
```
