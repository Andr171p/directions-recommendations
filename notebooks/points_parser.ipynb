```python
import pandas as pd
```

```python
df = pd.read_csv("ТИУ Бакалавриат Абитуриенты.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head(3)
```

```python
df.shape
```

```python
df = df[df["Вид приема"] == "Общий конкурс"]
df = df[df["Приказ о зачислении"] == 1]
df.shape
```

```python
df.head(3)
```

```python
points_by_per_year = df.groupby(['Год приёмной компании', 'Направление подготовки'])['Сумма баллов'].min()
```

```python
points_by_per_year
```

```python
points_by_per_year_df = points_by_per_year.reset_index(name='Минимальная сумма баллов')
points_by_per_year_df.head(3)
```

```python
points_by_per_year_df.to_csv("Проходные баллы 2019-2024.csv")
```

```python
directions_df = pd.read_csv("Applicants_With_Direction.csv")
directions_df.drop("Unnamed: 0", axis=1, inplace=True)
directions_df = directions_df[["direction_name", "direction_id"]]
directions_df.head(3)
```

```python
directions_df = directions_df.drop_duplicates()
directions_df
```

```python
directions_df.shape
```

```python
directions_df = directions_df.dropna()
directions_df.shape
```

```python
directions_df.head(5)
```

```python
points_by_per_year_df = points_by_per_year_df.rename(
    {
        "Год приёмной компании": "year",
        "Направление подготовки": "direction_name",
        "Минимальная сумма баллов": "points"
     }, axis=1
)
points_by_per_year_df.head(5)
```

```python
merged_df = pd.merge(
    directions_df,
    points_by_per_year_df,
    how='left',
    on='direction_name'
)
merged_df.head(5)
```

```python
merged_df.shape
```

```python
merged_df.to_csv("Проходные баллы 2019-2024.csv")
```
