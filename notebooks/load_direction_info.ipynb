```python
import pandas as pd
```

```python
file_path = "ВИ (1).xlsx"
df = pd.read_excel(file_path)
df.head(5)
```

```python
structured_data = []
current_direction = None

for index, row in df.iterrows():
    # Если есть новая форма обучения и направление
    if pd.notna(row["Форма[1] обучения"]) and pd.notna(row["Шифр и наименование направления"]):
        if current_direction:
            structured_data.append(current_direction)
        
        current_direction = {
            "Форма обучения": row["Форма[1] обучения"],
            "Шифр и наименование направления": row["Шифр и наименование направления"],
            "Вступительные испытания": [],
            "Описание направления подготовки/специальности": row["Описание направления подготовки/специальности"]
        }
    
    # Добавление вступительных испытаний
    if pd.notna(row["Вступительные испытания"]) and row["Вступительные испытания"] != "Приоритет ВИ":
        current_direction["Вступительные испытания"].append({
            "Приоритет": row["Вступительные испытания"],
            "Наименование": row["Unnamed: 3"],
            "Минимальный балл": row["Минимальный балл"]
        })

# Добавляем последнее направление
if current_direction:
    structured_data.append(current_direction)

# Вывод структурированных данных
for direction in structured_data:
    print(direction)
```

```python
import json

with open("Структурированные_направления_подготовки.json", "w", encoding="utf-8") as f:
    json.dump(structured_data, f, ensure_ascii=False, indent=4)
```
