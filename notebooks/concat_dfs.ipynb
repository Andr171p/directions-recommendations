```python
import pandas as pd
```

```python
file_path = r"C:\Users\andre\IdeaProjects\TyuiuDirectionsRecSys\data\Подготовленные_данные\Проходные баллы 2019-2024.csv"

df_2019_2024 = pd.read_csv(file_path)
df_2019_2024.drop(["Unnamed: 0", "direction_id"], axis=1, inplace=True)
df_2019_2024 = df_2019_2024.rename(columns={"direction_name": "Направление подготовки", "year": "Год", "points": "Сумма баллов"})
df_2019_2024.head(3)
```

```python
direcrions_ids_df = pd.read_csv(r"C:\Users\andre\IdeaProjects\TyuiuDirectionsRecSys\datasets\tyuiu_unique_directions.csv")
direcrions_ids_df.head(3)
```

```python
def get_id_by_direction_name(direction_name):
    filtered_rows = direcrions_ids_df[direcrions_ids_df['Направление подготовки'] == direction_name]
    if len(filtered_rows) > 0:
        return filtered_rows['ID'].values[0]
    else:
        return None 
```

```python
df_2019_2024["ID"] = df_2019_2024["Направление подготовки"].apply(get_id_by_direction_name)
    
df_2019_2024.head(3)
```

```python
df_2019_2024.fillna(-1)
df_2019_2024
```

```python
df_2019_2024.to_csv("ТИУ_проходные_баллы_2019-2024.csv")
```

```python
df_2014 = pd.read_csv(r"C:\Users\andre\IdeaProjects\TyuiuDirectionsRecSys\notebooks\ТИУ_проходные_баллы_2014.csv")
df_2014.drop("Unnamed: 0", axis=1, inplace=True)
df_2014 = df_2014.rename(columns={"Зачислен на направление подготовки (специальность)": "Направление подготовки"})
df_2014["ID"] = df_2014["Направление подготовки"].apply(get_id_by_direction_name)

file_paths = [
    "ТИУ_проходные_баллы_2015.csv",
    "ТИУ_проходные_баллы_2016.csv",
    "ТИУ_проходные_баллы_2017.csv",
    "ТИУ_проходные_баллы_2018.csv"
]

dfs = []
dfs.append(df_2014)
for file_path in file_paths:
    df = pd.read_csv(file_path)
    df.drop("Unnamed: 0", axis=1, inplace=True)
    df = df.rename(columns={"Направление, специальность, профессия": "Направление подготовки", "Итоговая сумма баллов": "Сумма баллов"})
    df["ID"] = df["Направление подготовки"].apply(get_id_by_direction_name)
    dfs.append(df)
    
dfs.append(df_2019_2024)
```

```python
merged_df = pd.concat(dfs, ignore_index=True)
merged_df
```

```python
merged_df.fillna(-1)
```

```python
merged_df.to_csv("ТИУ_проходные_баллы_2014-2024.csv")
```
